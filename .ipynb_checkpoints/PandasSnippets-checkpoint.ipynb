### How to filter dataframe by rows with multicriteria

```python
movies = pd.read_csv("http://bit.ly/imdbratings")
```

```python
movies
```

```python
movies.to_csv("movies.csv", index = False)
```

```python
movies[(movies.duration >= 200) & (movies.genre == 'Drama')]
```

## Multiple OR conditions by using the isin() method.

```python
movies[movies.genre.isin(['Crime', 'Drama', 'Action'])]
```

### Only keep the columns that are number types in a dataframe.

```python
drinks = pd.read_csv("http://bit.ly/drinksbycountry")
```

```python
drinks
```

```python
drinks.to_csv("drinks.csv", index = False)
```

```python
import numpy as np
drinks.select_dtypes(include=[np.number]).dtypes
d = drinks.select_dtypes(include=[np.number]).dtypes
```

```python
number_drinks = drinks[d.index]
```

### Get the mean for each series 

```python
number_drinks.mean()
```

```python
# get the mean for each row
number_drinks.mean(axis = 1)
```

### Check for content in each row. 

```python
orders = pd.read_table("http://bit.ly/chiporders")
orders.head(2)
```

```python
orders.item_name.str.contains('Chicken')
```

```python
orders[orders.item_name.str.contains('Chicken')]
```

### How to aggregate by group and extract stats

```python
drinks.groupby("continent").beer_servings.agg(['count', 'min', 'max', 'mean'])
```

```python
%matplotlib inline
drinks.groupby("continent").beer_servings.mean().plot(kind = 'bar')
```
