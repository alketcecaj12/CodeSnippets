### Cross tabulation in Pandas

```python
import pandas as pd

data = {
    'Make': ['Toyota', 'Honda', 'Ford', 'Toyota', 'Honda', 'Ford', 'Toyota', 'Honda', 'Ford', 'Toyota'],
    'Model': ['Camry', 'Civic', 'F-150', 'Corolla', 'Accord', 'Mustang', 'RAV4', 'CR-V', 'Explorer', 'Highlander'],
    'Color': ['Red', 'Blue', 'Black', 'White', 'Silver', 'Red', 'Blue', 'White', 'Black', 'Silver'],
    'Sales': [5, 3, 4, 6, 4, 2, 7, 5, 3, 4]
}

df = pd.DataFrame(data)
print(df)
```

```python
# Create a crosstab of car makes and colors
crosstab_result = pd.crosstab(df['Make'], df['Color'])
print("Car Makes by Color:")
print(crosstab_result)
```

```python
# Create a crosstab of car makes and colors, with sales values
crosstab_sales = pd.crosstab(df['Make'], df['Color'], values=df['Sales'], aggfunc='sum')
print("\nTotal Sales by Make and Color:")
print(crosstab_sales)
```

### Another use for category data type 

```python
mydata = pd.DataFrame({"ID": [103, 104, 100, 102, 106], 
                      "quality":["excellent", "good", "good", "very good", "bad"]})
```

```python
mydata
```

```python
from pandas.api.types import CategoricalDtype
cat_type = CategoricalDtype(categories=[ "bad", "good",  "very good", "excellent"], ordered=True)
```

```python
mydata['quality'] = mydata.quality.astype(cat_type)
```

```python
mydata.quality
```

```python
mydata.sort_values(by = "quality", ascending=False)
```
